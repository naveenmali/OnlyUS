# Connecting to mysql

```python
def dfto_sql(host,user,passw,port,database,tablename,df):
    mydb = create_engine ('mysql+pymysql://' + user + ':' + passw + '@' + host + ':' + str (port) + '/' + database,
                          echo=False)
    df.to_sql (name=tablename, con=mydb, if_exists='replace', index=False)
    return True


def dfread_sql(host,user,passw,port,database,query):
    mydb = create_engine ('mysql+pymysql://' + user + ':' + passw + '@' + host + ':' + str (port) + '/' + database,
                          echo=False)
    
    frame  = pd.read_sql(query, mydb);
    return frame



```
